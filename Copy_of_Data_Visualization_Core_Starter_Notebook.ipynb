## Explore BMI

### 9\) Regplot - Charges vs. BMI

9a) Make a seaborn regplot to visualize the relationship between charges and bmi for the entire dataset

```python
# Make a seaborn regplot for bmi vs. charges.
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='bmi', y='charges', scatter_kws={'alpha':0.5})
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.title('Relationship between Charges and BMI')
plt.show()
```

9b\) What do you notice from the visualization? *(There are many correct answers.)*

- Your Answer:
  - `...`There is a generally positive linear trend between charges and BMI.
       While the trend is positive, there is significant variability in charges for each BMI value.

### 10\) Scatterplot - Charges vs. BMI - Colored by Smoker

10a) Make a scatterplot to visualize the relationship between charges and bmi colored by smoker or not

```python
# Make a scatterplot of bmi vs charges, colored by smoker
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='bmi', y='charges', hue='smoker', alpha=0.7)
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.title('Relationship between Charges and BMI, Colored by Smoker')
plt.legend(title='Smoker', labels=['Non-Smoker', 'Smoker'])
plt.show()
```

10b) What do you notice from the visualization? (There are many correct answers.)

- Your Answer:
  - `...`The plot shows that smokers (designated by one color) generally have higher charges compared to non-smokers (designated by another color), regardless of their BMI.

  Within both smoker and non-smoker groups, there is a positive linear trend between charges and BMI. As BMI increases, charges tend to increase.

### 11\)  Linear Model Plot - Charges vs. BMI - Colored by Smoker



11a) Compare the relationship of bmi and charges for smokers and nonsmokers using seaborn's `lmplot` to calculate a separate trendline for Smokers vs. Non-Smokers.


```python
# Create an lmplot of bmi vs. charges, colored by smoker.
plt.figure(figsize=(10, 6))
sns.lmplot(data=df, x='bmi', y='charges', hue='smoker', markers=['o', 'x'], scatter_kws={'alpha': 0.7})
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.title('Linear Relationship between BMI and Charges, Colored by Smoker')
plt.legend(title='Smoker', labels=['Non-Smoker', 'Smoker'])
plt.show()
```

11b) What do you notice from the visualization? (Compare the trendlines.)

- Your Answer:
  - `...`The trendline for smokers is significantly steeper compared to that of non-smokers. This indicates that the increase in charges per unit increase in BMI is much more pronounced for smokers than for non-smokers.

  The steeper slope of the trendline for smokers suggests a stronger positive correlation between BMI and charges among smokers. In other words, for smokers, an increase in BMI is associated with a larger increase in charges compared to non-smokers

### 14\) Bar Chart of Ratios (%) of Smokers in Each Region

14a) Calculate the ratio of smokers in each region and save it as a variable called "percent_smoker_by_region". (Hint: use .groupby with "region" combined with slicing the "smoker" column and end with .mean() )

```python
percent_smoker_by_region = df.groupby('region')['smoker'].mean()

print(percent_smoker_by_region)
```

14b) Plot the ratios of smokers by region as a bar plot.



```python
# Plot the ratios (the results of then groupby above)
percent_smoker_by_region = df.groupby('region')['smoker'].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=percent_smoker_by_region.index, y=percent_smoker_by_region.values)
plt.xlabel('Region')
plt.ylabel('Percentage of Smokers')
plt.title('Percentage of Smokers by Region')
plt.show()
```
